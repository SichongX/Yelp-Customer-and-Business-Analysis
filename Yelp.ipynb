```python
# Give a socre(1-4) to each user based on the average populatiry of each review and compliments received
x = 'review_popularity'
y = 'review_popularity_score'
for i in user_cleaned.index:
    if user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.25):
        user_cleaned.loc[i,y] = 1
    if user_cleaned[x].quantile(q = 0.25) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.5):
        user_cleaned.loc[i,y] = 2
    if user_cleaned[x].quantile(q = 0.5) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.75):
        user_cleaned.loc[i,y] = 3
    if user_cleaned[x].quantile(q = 0.75) < user_cleaned.loc[i,x]:
        user_cleaned.loc[i,y] = 4
```

```python
x = 'compliments'
y = 'compliments_score'
for i in user_cleaned.index:
    if user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.25):
        user_cleaned.loc[i,y] = 1
    if user_cleaned[x].quantile(q = 0.25) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.5):
        user_cleaned.loc[i,y] = 2
    if user_cleaned[x].quantile(q = 0.5) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.75):
        user_cleaned.loc[i,y] = 3
    if user_cleaned[x].quantile(q = 0.75) < user_cleaned.loc[i,x]:
        user_cleaned.loc[i,y] = 4
```

```python
user_cleaned['user_score']= user_cleaned['review_popularity_score']+ user_cleaned['compliments_score']
user_cleaned.count()
```

```python
# Give each user a tier based on which quantile the user in
x = 'user_score'
y = 'user_tier'
for i in user_cleaned.index:
    if user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.25):
        user_cleaned.loc[i,y] = 'Level 1'
    if user_cleaned[x].quantile(q = 0.25) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.5):
        user_cleaned.loc[i,y] = 'Level 2'
    if user_cleaned[x].quantile(q = 0.5) < user_cleaned.loc[i,x] <= user_cleaned[x].quantile(q = 0.75):
        user_cleaned.loc[i,y] = 'Level 3'
    if user_cleaned[x].quantile(q = 0.75) < user_cleaned.loc[i,x]:
        user_cleaned.loc[i,y] = 'Level 4'
```

```python
x = 'user_tier'
y = 'user_weight'
for i in user_cleaned.index:
    if user_cleaned.loc[i,x] == 'Level 1':
        user_cleaned.loc[i,y] = 0.2
    if user_cleaned.loc[i,x] == 'Level 2':
        user_cleaned.loc[i,y] = 0.23
    if user_cleaned.loc[i,x] == 'Level 3':
        user_cleaned.loc[i,y] = 0.27
    if user_cleaned.loc[i,x]== "Level 4":
        user_cleaned.loc[i,y] = 0.3
```

```python
# Define a new founction to count the weighted averge stars by combining all reviews of a business
def w_avg (df, stars_review, user_weight):
    stars = df[stars_review] 
    weight = df[stars_review]
    return (stars * weight).sum()/ weight.sum()
```

```python
weighted_stars= review_user_business.groupby(by = 'business_id', as_index = False).apply(w_avg, 'stars_review', 'user_weight')
weighted_stars
```

```python
weighted_stars = weighted_stars.rename(columns = str).rename(columns = {'None': 'weighted_stars'})
weighted_stars['weighted_stars'] = weighted_stars['weighted_stars'].apply(lambda x: round(x,1))
weighted_stars
```

```python
new_stars = weighted_stars.merge(business_cleaned[['business_id','stars']], how = 'left')
new_stars
```

```python
# Calculate the mean of the stars displayed by Yelp and weighted stars as a new star for each business 
new_stars['new_stars'] = (new_stars['weighted_stars'] + new_stars['stars'])/2
new_stars['new_stars'] = new_stars['new_stars'].apply(lambda x: round(x,1))
new_stars.drop(['stars'], axis = 1, inplace = True)
new_stars
```

```python
new_business = business_cleaned.merge(new_stars, how = 'right')
new_business
```
